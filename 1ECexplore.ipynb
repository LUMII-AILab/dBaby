```python
""" Trains an agent with Episodic Memory control on Pong. Uses OpenAI Gym. """
import numpy as np
import _pickle as pickle
import gzip
import gym
import random
import npickle
%matplotlib inline
import matplotlib.pyplot as plt
import types
import torch

def _step_custom(self, a):
    env = self.unwrapped or self
    action = env._action_set[a]
    num_steps = 4 #self.np_random.randint(2, 5)
    ob = []; reward = 0.0
    for _ in range(num_steps):
        reward += env.ale.act(action)
        ob.append(env._get_obs())
    ob = np.maximum.reduce(ob) # returns max pixel values from 4 frames
    return ob, reward, env.ale.game_over(), {"ale.lives": env.ale.lives()} # lives left



H = 200 # number of hidden layer neurons
batch_size = 10 # every how many episodes to do a param update?
learning_rate = 1e-4
gamma = 0.99 # discount factor for reward
decay_rate = 0.99 # decay factor for RMSProp leaky sum of grad^2
explore = 0.005
dimensions = 64
kkk = 11
start = 10
seed = 332
D = 105 * 80 # input dimensionality: 80x80 grid

# gamespace 
#env = gym.make("SpaceInvaders-v3") # environment info
env = gym.make("Pong-v3") # environment info
#env = gym.make("Breakout-v3") # environment info
#env = gym.make("MontezumaRevenge-v3") # environment info

# env = gym.make("CrazyClimber-v3") # environment info
# env = gym.make("Asteroids-v3") # environment info
# env = gym.make("Boxing-v3")
# env = gym.make("Pong-v3")
# env = gym.make("Bowling-v3")
# env = gym.make("Seaquest-v3")
# env = gym.make("Robotank-v3")
# env = gym.make("MontezumaRevenge-v3")
# env = gym.make("Frostbite-v3")

env.seed(seed)
torch.manual_seed(seed)
np.random.seed(seed)
random.seed(seed)
env._step = types.MethodType(_step_custom, env)
print ( 'action count:', env.action_space.n)
valid_actions = range(env.action_space.n)

running_reward = None
lgraphx, lgraphy = [], []
memoryAct, memoryScr = [], []
memorySta, memoryRew, memoryImg, memoryImgY = [], [], [], []
for i in range(len(valid_actions)):
            memorySta.append([])
            memoryRew.append([])
            memoryImg.append([])
            memoryImgY.append([])
rmatrix = np.random.normal(0,1.0, (D,dimensions)) # One dimension/hash for random
    
def prepro(I):
  """ prepro 210x160x3 uint8 frame into 6400 (80x80) 1D float vector """
  I = I[::2,::2,2] # downsample by factor of 2
  I[I == 17] = 0 # erase background (background type 1)
  I[I == 192] = 0 # erase background (background type 1)
  I[I == 136] = 0 # erase background (background type 1)
  I[I != 0] = 1 # everything else (paddles, ball) just set to 1
#  plt.imshow(I) 
#  plt.show()  
  return I.astype(np.float).ravel()

def discount_rewards(r):
  """ take 1D float array of rewards and compute discounted reward """
  discounted_r = np.zeros_like(r)
  running_add = 0
  for t in reversed(range(0, r.size)):
    #if r[t] != 0: running_add = 0 # reset the sum, since this was a game boundary (pong specific!)
    # For SpaceInvaders EM exploration to progress, this needs to be disabled
    running_add = running_add * gamma + r[t]
    discounted_r[t] = running_add
  return discounted_r

def discount_rewards_Pong(r):
  """ take 1D float array of rewards and compute discounted reward """
  discounted_r = np.zeros_like(r)
  running_add = 0
  for t in reversed(range(0, r.size)):
    if r[t] != 0: running_add = 0 # reset the sum, since this was a game boundary (pong specific!)
    # For SpaceInvaders EM exploration to progress, this needs to be disabled
    running_add = running_add * gamma + r[t]
    discounted_r[t] = running_add
  return discounted_r

def code(i, in_state, c_state, kk): # KNN
    diffs = c_state-in_state
    squared_diffs = diffs * diffs
    squared_distances = torch.sum(squared_diffs, 1)
    values, indices_of_closest_distances = torch.topk(-squared_distances, k=kk) #, sorted=True)
    return indices_of_closest_distances, torch.sqrt(-values)

def simi(i, in_state, c_state, kk): # KNN
    #cosi = torch.mm(c_state,in_state.t)
    #print (in_state, c_state)
    cosi = c_state.mm(in_state.unsqueeze(0).t())
    #cosi = in_state.mm(c_state) #.unsqueeze(0))
    #print (cosi)
    values, indices_of_closest_distances = torch.topk(cosi.t(), k=kk) #, sorted=True)
    #val = torch.sqrt(2 - 2 * values)
    #print (values, indices_of_closest_distances)
    return indices_of_closest_distances[0].cpu().numpy().T, values[0].cpu().numpy().T

    
for episode_number in range(167): #805 : # True:
    observation = env.reset()
    prev_x = None # used in computing the difference frame
    xs,ys,ss,dlogsp,fa,drs = [],[],[],[],[],[]
    reward_sum = 0
    if episode_number > start :
        c_states = []
        for i in range(len(valid_actions)):    
            c_states.append(torch.from_numpy(np.array(memorySta[i])).float().cuda())

    for steps in range(10000):
        cur_x = prepro(observation)   
        x = cur_x - prev_x if prev_x is not None else np.zeros(D)   
        prev_x = cur_x
        state2 = np.dot(x.reshape(-1), rmatrix)
        #state2 = x.reshape(-1)
        
        sume = np.sqrt(np.dot(state2,state2.T))
        if (sume == 0) :
              state = state2
        else:
            state = state2 / sume

        if (np.random.uniform(0,1) > explore) and (episode_number > start):
            distances, inds, mu = [], [], [] 
            for i in range(len(valid_actions)):
                in_state = torch.from_numpy(state).float().cuda()
                ind, di = code(i, in_state, c_states[i], kkk)
                inds.append(ind)
                mu.append(0)
                for z in range (kkk):
                    mu[i] += memoryRew[i][inds[i][z]] / kkk

            mumu = max(mu)
            yyys =[]
            for z in range(len(valid_actions)):
                if mu[z] == mumu:
                    yyys.append(z)
            yyy = random.choice(yyys)

            aprob = valid_actions[yyy]

        else: aprob = valid_actions[random.choice(valid_actions)]

        xs.append(x) # observation
        ss.append(state) # observation

        dlogsp.append(aprob)
        observation, reward, done, info = env.step(aprob)
        reward_sum += reward
        drs.append(reward) # record reward (has to be done after we call step() to get reward for previous action)

        cur_xx = prepro(observation)   
        xx = cur_xx - prev_x if prev_x is not None else np.zeros(D)   
        ys.append(xx)

        if done:
            break


    # stack together all inputs, hidden states, action gradients, and rewards for this episode
    epr = np.vstack(drs)
    discounted_epr = discount_rewards(epr)
    discounted_epr_Pong = discount_rewards_Pong(epr)


    for k in range(len(discounted_epr)):

                memorySta[dlogsp[k]].append(ss[k].tolist())
                memoryRew[dlogsp[k]].append(discounted_epr[k,0])
                memoryImg[dlogsp[k]].append(xs[k])
                memoryImgY[dlogsp[k]].append(ys[k])
                if discounted_epr_Pong[k,0] > 0 : # si bija mana pus-gada kluda!!!
                    memoryScr.append(xs[k])
                    oneHotAction = np.zeros_like(valid_actions)
                    oneHotAction[dlogsp[k]] = 1.0
                    memoryAct.append(oneHotAction)

    
    # boring book-keeping
    accumulate = min([episode_number, 10])
    running_reward = reward_sum if running_reward is None else running_reward * (1-1/accumulate) + reward_sum * (1/accumulate)
    print ('episode: {}, reward: {}, mean reward: {:3f}'.format(episode_number, reward_sum, running_reward))
    lgraphx.append(episode_number)
    lgraphy.append(running_reward)
    
    if episode_number % 20 == 0:  
        plt.plot(lgraphx, lgraphy)
        plt.show() 

        npickle.dump_gzip(memoryImg, 'ZmemoryImgX.k')
        npickle.dump_gzip(memoryImgY, 'ZmemoryImgY.k')
        npickle.dump_gzip(memoryRew, 'ZmemoryRew.k')
        
        npickle.dump_gzip(memoryScr, 'XmemoryScrZ.k')
        npickle.dump_gzip(memoryAct, 'XmemoryActZ.k')
        print ('chekpoint saved',len(memoryImg[0]), len(memoryImgY[0]), len(memoryRew[0]), len(memoryScr), len(memoryAct))

```
